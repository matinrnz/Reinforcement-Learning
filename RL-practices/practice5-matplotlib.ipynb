### implementing figuers and axis

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATA
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
fig = plt.figure()

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')


# SMALL AXES
axes2 = fig.add_axes([1,1,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```

```python
type(fig)
```

```python
fig = plt.figure(figsize=(2,2),dpi=100)

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

fig.savefig('new_figure.jpg',bbox_inches='tight')
```
