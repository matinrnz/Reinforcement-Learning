### visual styling

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x,color='#d1ce1d',lw=10,ls=':') # RGB HEX Code

lines = ax.plot(x,x+1,color='purple',linewidth=2,marker='s',ls='--',ms=10,markerfacecolor='red',markeredgewidth='4',markeredgecolor='orange')
# lines[0].set_dashes([1,2,1,2,10,2])
# ax.plot(x,x+1,color='purple',label="x vs x+1")

# lines = ax.plot(x,x,color='purple',linewidth=2,marker='+ o',ls='--',markersize=20)

ax.legend()
```
